# 2. Data Cleaning

```python
# Remove duplicate records
df.drop_duplicates(inplace=True)
df
```

```python
#Handle missing values appropriately

# removing unwanted columns-url,phone
df = df.drop(["url","phone","Unnamed: 0","address","location","reviews_list","menu_item","rest_type"],axis = 1)

```

```python
# Renaming columns
d = {"listed_in(city)" : "location","listed_in(type)" : "rest_type","approx_cost(for two people)" : "cost_for_two"}
df.rename(columns = d,inplace = True)
```

```python
# handling missing values in rate column and changing datatype
df['rate'] = df['rate'].replace("NEW",np.nan)
df['rate'] = df['rate'].replace("-",np.nan)
df['rate'] = df['rate'].str.strip("/5")
df['rate'] = df['rate'].astype(float)
df['rate'] = df['rate'].fillna(round(df['rate'].mean()))
```

```python
df['cost_for_two'].unique()
df['cost_for_two'] = df['cost_for_two'].str.replace(",","")
df['cost_for_two'] = df['cost_for_two'].astype(float)
# # df['approx_cost(for two people)'].mode()
df['cost_for_two'] = df['cost_for_two'].fillna(round(df['cost_for_two'].mean()))
df['cost_for_two'] = df['cost_for_two'].astype(int)
```

```python
df['dish_liked'].mode()
df['dish_liked'].unique()
df['dish_liked'] = df['dish_liked'].fillna(df['dish_liked'].mode()[0])
```

```python
df['cuisines'].mode()
df['cuisines'].unique()
df['cuisines'] = df['cuisines'].fillna(df['cuisines'].mode()[0])
```

# Task 2: Univariate Analysis

```python
# importing libraries for data visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Top 5 locations with highest number of restaurants

top_locations = df['location'].value_counts().head(5)

plt.figure()
plt.bar(top_locations.index, top_locations.values, color="red", width=0.3)
plt.title("Top 5 Locations with Highest Number of Restaurants")
plt.xlabel("Location")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=90)
plt.show()
```

1. BTM , Koramangala 7th Block , Koramangala 5th Block , Koramangala 4th Block, Koramangala 6th Block are the top 5 locations with highest no. of restaurants

2. customer demand could be more

```python
# Distribution of Restaurant Ratings

sns.histplot(x="rate", data=df, kde=True)
plt.title("Distribution of Restaurant Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```



1.   Most of the restaurants rating lie between 3.5 to 4.5
2.   very few restaurants have low or high ratings

```python
#Percentage of Restaurants Offering Online Orders

online_orders = df['online_order'].value_counts()

plt.figure()
plt.pie(online_orders, labels=online_orders.index, autopct='%0.0f%%')
plt.title("Percentage of Restaurants Offering Online Orders")
plt.show()
```

```python
#Distribution of average cost for two

average=df.groupby('location')['cost_for_two'].mean()

sns.histplot(x=average,kde=True,color="red")
plt.title("Distribution of Average Cost for Two")
plt.xlabel("Average Cost")
plt.ylabel("Frequency")
```



1.   It is right skewed showing most restaurants have moderate
price range

## Task 3: Bivariate Analysis


```python
# Rating vs Cost

plt.scatter(df['cost_for_two'], df['rate'])

# titles
plt.title("Rating vs Cost for Two")
plt.xlabel("Cost for Two")
plt.ylabel("Rating")

plt.show()
```



1.   Rating increases as cost increases
2.   Restaurants can focus on quality of food



```python
# Group by online_order and calculate average rating
data = df.groupby('online_order')['rate'].mean().reset_index()

plt.figure(figsize=(6,5))

# line graph
plt.plot(data['online_order'], data['rate'], marker='o', color='red')

# title and labels
plt.title("Average Rating vs Online Order")
plt.xlabel("Online Order Availability")
plt.ylabel("Average Rating")


plt.show()
```

```python
# Calculating average rating for Yes and No
avg_rating = df.groupby('online_order')['rate'].sum()

plt.figure(figsize=(6,5))

# bar chart
plt.bar(avg_rating.index, avg_rating.values)

# title and labels
plt.title("Rating by Online Order Availability")
plt.xlabel("Online Order (Yes/No)")
plt.ylabel("Rating")

plt.show()
```



*  Compared to offline, availability of online order gains much rating




```python
# Group by online_order and calculate average rating
data = df.groupby('online_order')['rate'].mean().reset_index()

# Create figure
plt.figure(figsize=(6,5))

# bar graph
plt.bar(data['online_order'], data['rate'], color='red')

#  title and labels
plt.title("Average Rating vs Online Order")
plt.xlabel("Online Order Availability")
plt.ylabel("Average Rating")

plt.show()
```

* Availability of online order doesn't has impact on average rating
* Both are same






```python
# Create a new figure with specified size
plt.figure(figsize=(8,5))

# scatter plot to show relationship between votes and rating
sns.scatterplot(x='votes', y='rate', data=df, color='red')

# Add title
plt.title("Votes vs Rating")

# Label axes
plt.xlabel("Votes")
plt.ylabel("Rating")


plt.show()
```



*   No. of votes doesn't have impact on rating wheter it is low or high
*   It is average of voters who gave rating so even the count is less, rating would be more
